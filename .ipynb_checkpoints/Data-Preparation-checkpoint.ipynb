```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
np.random.seed(42)
```

```python
def generate_features(n_samples=1500):

    data = {
        "HR_mean": np.random.normal(75, 10, n_samples),
        "HR_std": np.random.normal(5, 2, n_samples),
        "RR_mean": np.random.normal(0.8, 0.1, n_samples),
        "RR_rmssd": np.random.normal(30, 8, n_samples),
        "GSR_mean": np.random.normal(0.5, 0.2, n_samples),
        "GSR_std": np.random.normal(0.05, 0.01, n_samples),
        "TEMP_mean": np.random.normal(36.5, 0.3, n_samples),
        "ACC_mean": np.random.normal(1.0, 0.2, n_samples),
        "ACC_std": np.random.normal(0.3, 0.1, n_samples),
        "Tap_speed": np.random.normal(3, 1, n_samples)
    }

    return pd.DataFrame(data)
```

```python
def feature_augmentation(df):

    augmented = df.copy()

    # Nonlinear interactions
    augmented["HR_GSR"] = df["HR_mean"] * df["GSR_mean"]
    augmented["RR_HR_ratio"] = df["RR_mean"] / df["HR_mean"]
    augmented["ACC_TEMP"] = df["ACC_mean"] * df["TEMP_mean"]

    # Polynomial terms
    augmented["HR_sq"] = df["HR_mean"] ** 2
    augmented["GSR_sq"] = df["GSR_mean"] ** 2

    # PCA latent features
    pca = PCA(n_components=2)
    pca_features = pca.fit_transform(df)

    augmented["PCA_1"] = pca_features[:, 0]
    augmented["PCA_2"] = pca_features[:, 1]

    return augmented
```

```python
wafd_datasets = []
afd_datasets = []

for i in range(8):

    # Generate base dataset
    base_df = generate_features(n_samples=1500)

    # WAFD
    wafd = base_df.copy()
    wafd_datasets.append(wafd)

    # AFD
    afd = feature_augmentation(base_df)
    afd_datasets.append(afd)

    print(f"Created Dataset {i+1} → WAFD & AFD")
```

```python
final_WAFD = pd.concat(wafd_datasets, axis=0).reset_index(drop=True)
```

```python
final_AFD = pd.concat(afd_datasets, axis=0).reset_index(drop=True)
```

```python
final_WAFD.to_csv("dataset/WAFD_Dataset.csv", index=False)
final_AFD.to_csv("dataset/AFD_Dataset.csv", index=False)
```
